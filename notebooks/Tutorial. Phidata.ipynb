#Importing Libraries

```python
import os
from google.colab import userdata
from phi.assistant import Assistant
from phi.llm.openai import OpenAIChat
from phi.tools.yfinance import YFinanceTools
from openai import OpenAI
from portkey_ai import PORTKEY_GATEWAY_URL, createHeaders
from phi.assistant.duckdb import DuckDbAssistant
from phi.tools.duckduckgo import DuckDuckGo
from portkey_ai import Portkey

def set_api_keys(default_openai_key: str = "YOUR_API_KEY", default_portkey_key: str = "YOUR_API_KEY") -> None:
    """Set the OpenAI API key from Google Colab's user environment or use a default value."""

    os.environ["OPENAI_API_KEY"] = userdata.get('TT_OPENAI_KEY') or default_openai_key
    os.environ["PORTKEY_API_KEY"] = userdata.get('PORTKEY_API_KEY') or default_portkey_key

set_api_keys()
```

# Creating A Basic Assistant Using Phidata

```python
# Initialize the OpenAIChat model
llm = OpenAIChat(
    model = "gpt-4o-mini",
    base_url=PORTKEY_GATEWAY_URL,
    api_key=os.environ["OPENAI_API_KEY"], #Replace with Your OpenAI Key
    default_headers=createHeaders(
        provider="openai",
        api_key= os.environ["PORTKEY_API_KEY"]  # Replace with your Portkey API key
    )
)


internet_agent = Assistant(
     llm=llm,
    tools=[DuckDuckGo()], show_tool_calls=True)

# # Use the assistant to print the response to the query "What is today?"
internet_agent.print_response("What is portkey AI")
```

# Assistant that can query financial data


```python
# Initialize the OpenAIChat model
llm = OpenAIChat(
    model = "gpt-4o-mini",
    base_url=PORTKEY_GATEWAY_URL,
    default_headers=createHeaders(
        provider="openai",
        api_key=userdata.get('PORTKEY_API_KEY')
    )
)


Stock_agent = Assistant(
    llm=llm,
    tools=[YFinanceTools(stock_price=True, analyst_recommendations=True, company_info=True, company_news=True)],
    show_tool_calls=True,
    markdown=True,


     )

Stock_agent.print_response("Write a comparison between MSFT, META and GOOGL, use all tools available.")
```
